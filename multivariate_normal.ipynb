```python
import numpy as np
import matplotlib.pyplot as plt

# mean [x, y]
mu0 = [-10, 0]
mu1 = [10, 0]

# Identity matrix
cov = np.asarray([[1, .9], [.9, 1]])

X_class0 = np.random.multivariate_normal(mu0, cov, 100)
X_class1 = np.random.multivariate_normal(mu1, cov, 100)

plt.scatter(X_class0[:, 0], X_class0[:, 1])
plt.scatter(X_class1[:, 0], X_class1[:, 1])
plt.show()
```
